```python
import navis
import flybrains
from fafbseg import flywire
```

```python
# Set ID 
root_id = [720575940636656632] # Web 
```

```python
# Get meshes 
n = flywire.get_mesh_neuron(root_id)
```

```python
# Get brain and neuropils 
brain = flybrains.FLYWIRE
```

```python
# Set colors 
color = []
color.append([204/255,51/255,153/255,1]) # Neuron
color.append([0.5,0.5,0.5,0.1]) # Brain
```

```python
# Plot neuron and brain
fig, ax = navis.plot2d(
    [n, brain.mesh],  
    dps_scale_vec = 1000, 
    method = '2d', 
    view = ('-x','-y'), # Posterior view: ('-x','-y'); top view: ('-x','-z')
    figsize = (12, 12),
    color = color
    )
ax.set_axis_off()
```
